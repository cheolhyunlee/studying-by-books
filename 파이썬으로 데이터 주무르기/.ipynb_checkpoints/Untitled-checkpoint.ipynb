```python
import pandas as pd
```

```python
CCTV_Seoul = pd.read_csv('CCTV_in_Seoul.csv',encoding = 'utf-8')
CCTV_Seoul.head()
```

```python
CCTV_Seoul.columns
```

```python
CCTV_Seoul.columns[0]
```

```python
CCTV_Seoul.rename(columns={CCTV_Seoul.columns[0]:'구별'}, inplace=True)
```

```python
CCTV_Seoul.head()
```

```python
Population_in_Seoul = pd.read_excel('Population_in_Seoul.xls',encoding = 'utf-8')
Population_in_Seoul.head()
```

```python
Population_in_Seoul = pd.read_excel('Population_in_Seoul.xls',header = 2,parse_cols='B,D,G,J,N',encoding = 'utf-8')
Population_in_Seoul.head()
```

```python
Population_in_Seoul.rename(columns = {Population_in_Seoul.columns[0]:'구별',
                                      Population_in_Seoul.columns[1]:'인구수',
                                      Population_in_Seoul.columns[2]:'한국인',
                                      Population_in_Seoul.columns[3]:'외국인',
                                      Population_in_Seoul.columns[4]:'고령자'}, inplace = True)
Population_in_Seoul.head()
```

```python
CCTV_Seoul.head()
```

```python
CCTV_Seoul.sort_values(by='소계',ascending = False)
```

```python
CCTV_Seoul['최근 증가율'] = ( CCTV_Seoul['2016년'] + CCTV_Seoul['2015년'] + CCTV_Seoul['2014년'] ) / CCTV_Seoul['2013년도 이전']*100
CCTV_Seoul.sort_values(by = '최근 증가율', ascending = False)                        
```

```python
Population_in_Seoul
```

```python
Population_in_Seoul = Population_in_Seoul.drop([0],inplace =False)
```

```python
Population_in_Seoul['구별'].unique()
```

```python
Population_in_Seoul['외국인비율'] = Population_in_Seoul['외국인'] / Population_in_Seoul['인구수'] * 100
Population_in_Seoul['고령자비율'] = Population_in_Seoul['고령자'] / Population_in_Seoul['인구수'] * 100
Population_in_Seoul
```

```python
Population_in_Seoul.sort_values(by = '인구수',ascending = False)
```

```python
Population_in_Seoul.sort_values(by = '외국인',ascending = False)
```

```python
Population_in_Seoul.sort_values(by = '외국인비율',ascending = False)
```

```python
Population_in_Seoul.sort_values(by = '고령자',ascending = False)
```

```python
Population_in_Seoul.sort_values(by = '고령자비율',ascending = False)
```

```python
data_result = pd.merge(CCTV_Seoul,Population_in_Seoul, on='구별')
data_result
```

```python
del data_result['2013년도 이전']
del data_result['2014년']
del data_result['2015년']
del data_result['2016년']
data_result
```

```python
data_result.set_index('구별')
```
