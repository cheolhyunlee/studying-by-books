```python
import urllib.request
import datetime
import time
import json
from config import *
import math
import requests
from bs4 import BeautifulSoup
import folium
```

```python
def get_request_url(url):
    
    req = urllib.request.Request(url)
    
    try: 
        response = urllib.request.urlopen(req)
        if response.getcode() == 200:
            return response.read().decode('utf-8')
    except Exception as e:
        print(e)
        print("[%s] Error for URL : %s" % (datetime.datetime.now(), url))
        return None
```

```python
#[CODE 1]

def SanggaInfoService(pageNo,numOfRows,sigunguCd,bjdongCd):
    
    end_point = "http://api.suwon.go.kr/openapi-data/service/SanggaInfo/getHjdsearchInfo"
    
    parameters = "?pageNo=" + str(pageNo)
    parameters += "&numOfRows=" + str(numOfRows)
    parameters += "&sigunguCd=" + str(sigunguCd)
    parameters += "&bjdongCd=" + str(bjdongCd)
    parameters += "&ServiceKey=" + "TIyuV2FHFQSS46wIlsJNrwYGs9SqQEDOtNb0XZjVWp8k6pVUitI9BsXYzNGEYhUmLuovIzoeoIEA3fbXwqQwTg%3D%3D"
    
    global url
    
    url = end_point + parameters
    
    global retData
    
    retData = get_request_url(url)
    
    if (retData == None):
        return None
    else:
        return 
```

```python
gu_code = 41117 
bjdongCd_suwon = [10100,10200,10300,10400,10500,10600,10700]
bjdong_suwon = ['매탄동','원천동','이의동','하동','영통동','신동','망포동']
```

```python
# 카페

yeongtong_cafe_lat = []
yeongtong_cafe_lon = []

for i in bjdongCd_suwon :
    
    SanggaInfoService(1,5000,gu_code,i)

    soup = BeautifulSoup(retData, "html.parser")
    items = soup.find_all("item")

    for i in items :
        if str(i.indsmclsnm.text) == "커피점/카페" :
            yeongtong_cafe_lat.append(i.lat.text)
            yeongtong_cafe_lon.append(i.lon.text)
```

```python
# 치킨

yeongtong_chi_lat = []
yeongtong_chi_lon = []

for i in bjdongCd_suwon :
    
    SanggaInfoService(1,5000,gu_code,i)

    soup = BeautifulSoup(retData, "html.parser")
    items = soup.find_all("item")

    for i in items :
        try:
            if str(i.ksicnm.text) == "치킨 전문점" :
                yeongtong_chi_lat.append(float(i.lat.text))
                yeongtong_chi_lon.append(float(i.lon.text))
        except:
            None
```

```python
# 편의점

con_info = []
yeongtong_con_store_lat = []
yeongtong_con_store_lon = []

for i in bjdongCd_suwon :
    
    SanggaInfoService(1,5000,gu_code,i)

    soup = BeautifulSoup(retData, "html.parser")
    items = soup.find_all("item")

    for i in items :
        try:
            if str(i.ksicnm.text) == "체인화 편의점" :
                con_info.append(i)
            
            if str(i.ksicnm.text) == "체인화 편의점" :
                yeongtong_con_store_lat.append(float(i.lat.text))
                yeongtong_con_store_lon.append(float(i.lon.text))
        except:
            None
```
