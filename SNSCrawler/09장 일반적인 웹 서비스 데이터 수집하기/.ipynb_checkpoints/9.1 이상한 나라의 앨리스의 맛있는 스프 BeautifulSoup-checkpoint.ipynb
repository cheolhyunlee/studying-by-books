```python
import urllib.request
from bs4 import BeautifulSoup
html = urllib.request.urlopen('http://movie.naver.com/movie/sdb/rank/rmovie.nhn')
soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify())
```

```python
tags = soup.findAll('div', attrs={'class':'tit3'})
tags
```

```python
for tag in tags:
	print(tag.a)
```

```python
for tag in tags:
	print(tag.a.text)
```
