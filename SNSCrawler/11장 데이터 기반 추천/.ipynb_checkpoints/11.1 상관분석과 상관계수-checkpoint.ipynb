```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import matplotlib
from jupyterthemes import jtplot
jtplot.style(theme = "oceans16")
```

```python
font_location = "c:/Windows/fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family=font_name)
height = [100, 120, 130, 140, 150, 160, 170, 180, 190]
foot_size = [200, 205, 210, 220, 230, 250, 270, 280, 285]
plt.scatter(height, foot_size, c = 'w')
plt.xlabel('키(Cm)')
plt.ylabel('발크기(mm)')
plt.show()
```

```python
height = [100, 200, 300, 400, 500, 600, 700, 800, 900]
temperature = [18.0, 17.5, 17, 16.5, 15, 14.5, 13, 12, 11]
plt.scatter(height, temperature, c = 'w')
plt.xlabel('산의높이(m)')
plt.ylabel('온도(℃)')
plt.show()
```
